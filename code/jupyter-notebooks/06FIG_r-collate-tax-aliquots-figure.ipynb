```r
pacman::p_load(ggplot2,tidyr,ggpattern,patchwork)
```

```r
for_plot=read.csv("/vortexfs1/scratch/akrinos/2022-euk-diversity-temp/bats-tax-aliquots.csv") %>%
    dplyr::rename(c("BLASTLCA"="Has.BLAST.LCA"))
```

```r
bats_plot <- ggplot(for_plot %>% dplyr::group_by(type_cat,BLASTLCA,Tax.aliquots.cat) %>%
    dplyr::summarize(transcript_name = sum(transcript_name))) + #geom_bar(aes(x=type_cat,y=transcript_name,fill=Tax.aliquots.cat),stat="identity") +
    xlab("") + 
  geom_col_pattern(
    aes(x=type_cat,y=transcript_name,pattern = BLASTLCA,
        fill = Tax.aliquots.cat), # <- make this an interaction
    colour = "black",
    pattern_fill = "black",
    pattern_angle = 45,
    pattern_density = 0.05,
    pattern_spacing = 0.05
  )+
    scale_alpha_manual(values=c(0.5,1),
                        breaks=c("Yes","No"),
                        labels=c("Recovered by BLAST-LCA",
                                "Only found by tax-aliquots"),
                        name="")+
    ylab("Number of sequences") +
    scale_fill_manual(values=c("#648FFF","#DC267F",
                                       "#FE6100","#FFB000","gray"),
                      breaks=c("Coherent","Disagrees","Mixed","Mixed eukaryota",
                               "None"),
                      labels=c("Contains only taxon of interest",
                               "Does not contain BLAST-LCA\ntaxon annotation",
                               "Contains both eukaryotic\nand prokaryotic sequences",
                               "Only contains eukaryotic sequences",
                               "Not in a tax-aliquots cluster"),
                               name="Tax-aliquots category",
    guide = guide_legend(override.aes = list(pattern = "none")))+
    theme_classic(base_size=12)+
  scale_pattern_manual(name="",
    breaks=c("No","Yes"),
    labels=c("Only found by tax-aliquots",
        "Recovered by BLAST-LCA"),
    values = c("stripe","none"),
    guide = guide_legend(override.aes = list(fill = "grey70")) # <- make lighter
  ) 
bats_plot
```
